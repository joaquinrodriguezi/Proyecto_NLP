## Ejemplo reducido: inspección básica

```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd().resolve()
if not (PROJECT_ROOT / 'src').exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
RAW_DIR = PROJECT_ROOT / 'data' / 'raw'
INPUT_FILE = RAW_DIR / 'UFC_champion_Jon_Jones_retires_comments.csv'
df = pd.read_csv(INPUT_FILE, encoding='utf-8')
print('shape:', df.shape)
print(df.head(3).to_string(index=False))
```

## Ejemplo reducido: guardar una muestra

```python
from pathlib import Path
import pandas as pd
from datetime import datetime

PROJECT_ROOT = Path.cwd().resolve()
if not (PROJECT_ROOT / 'src').exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
RAW_DIR = PROJECT_ROOT / 'data' / 'raw'
ARTIFACTS_DIR = PROJECT_ROOT / 'artifacts'
ARTIFACTS_DIR.mkdir(exist_ok=True)
INPUT_FILE = RAW_DIR / 'UFC_champion_Jon_Jones_retires_comments.csv'
df = pd.read_csv(INPUT_FILE, encoding='utf-8')
sample = df.head(200).copy()
out_path = ARTIFACTS_DIR / f"load_inspect_sample_{datetime.now().strftime('%Y%m%d_%H%M%S')}.csv"
sample.to_csv(out_path, index=False, encoding='utf-8')
print('Muestra guardada en', out_path)
```
