## Qué hace el script

El script crea una submuestra reproducible con random_state=42, calcula métricas de longitud, detecta emojis/URLs/hashtags y genera tablas y gráficos de frecuencias. Su objetivo es convertir el corpus en un conjunto de hallazgos exploratorios claros antes de preprocesar y modelar.

## Ejemplo reducido: muestra y longitud

```python
from pathlib import Path
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

PROJECT_ROOT = Path.cwd().resolve()
if not (PROJECT_ROOT / 'src').exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
RAW_DIR = PROJECT_ROOT / 'data' / 'raw'
ARTIFACTS_DIR = PROJECT_ROOT / 'artifacts'
ARTIFACTS_DIR.mkdir(exist_ok=True)
INPUT_FILE = RAW_DIR / 'UFC_champion_Jon_Jones_retires_comments.csv'
df = pd.read_csv(INPUT_FILE, encoding='utf-8').sample(n=500, random_state=42)
df['n_chars'] = df['content'].fillna('').astype(str).str.len()
plt.figure(figsize=(6, 3))
sns.histplot(df['n_chars'], bins=20)
plt.title('Longitud de comentarios (muestra)')
plt.tight_layout()
out_path = ARTIFACTS_DIR / 'eda_sample_hist.png'
plt.savefig(out_path, dpi=150)
plt.close()
print('Histograma guardado en', out_path)
print(df[['content']].head(3).to_string(index=False))
```

## Ejemplo reducido: top tokens

```python
import re
from collections import Counter
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd().resolve()
if not (PROJECT_ROOT / 'src').exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
RAW_DIR = PROJECT_ROOT / 'data' / 'raw'
INPUT_FILE = RAW_DIR / 'UFC_champion_Jon_Jones_retires_comments.csv'
df = pd.read_csv(INPUT_FILE, encoding='utf-8').sample(n=500, random_state=42)
text = ' '.join(df['content'].fillna('').astype(str).tolist()).lower()
words = re.findall(r"[a-zA-Z']+", text)
print(Counter(words).most_common(10))
```
