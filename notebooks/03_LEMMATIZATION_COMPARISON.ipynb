## Ejemplo reducido: limpiar texto

```python
from pathlib import Path
import pandas as pd
import importlib.util

PROJECT_ROOT = Path.cwd().resolve()
if not (PROJECT_ROOT / 'src').exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
RAW_DIR = PROJECT_ROOT / 'data' / 'raw'
SAMPLE_FILE = RAW_DIR / 'sample_5000.csv'
module_path = PROJECT_ROOT / 'src' / 'comparacion_NLTK vs spaCy.py'
spec = importlib.util.spec_from_file_location('comparison_mod', module_path)
mod = importlib.util.module_from_spec(spec)
spec.loader.exec_module(mod)
df = pd.read_csv(SAMPLE_FILE, encoding='utf-8').head(50)
df['cleaned'] = df['content'].fillna('').astype(str).apply(mod.clean_text)
print(df[['content', 'cleaned']].head(5).to_string(index=False))
```
