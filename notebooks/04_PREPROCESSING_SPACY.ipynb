## Qué hace el script

El script define regex para limpiar URLs, menciones y hashtags, preserva emojis, aplica lematización con spaCy y guarda archivos procesados listos para vectorizar. La decisión de conservar emojis y reemplazarlos por etiquetas es clave para el rendimiento del pipeline.

## Funciones y clases usadas

- **clean_text**
- **emoji_to_label**
- **map_emojis**

## Gráficos y artefactos

- Se genera data/processed/sample_5000_clean.csv.
- Se genera data/processed/sample_5000_lemmatized.csv.
- Se guarda artifacts/emoji_map.txt y artifacts/clean_examples_before_after.csv.

## Ejemplo reducido: limpiar un texto corto
