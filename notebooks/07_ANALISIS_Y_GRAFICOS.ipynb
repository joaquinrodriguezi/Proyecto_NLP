## Ejemplo reducido: cargar métricas de entrenamiento

```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd().resolve()
if not (PROJECT_ROOT / 'src').exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
ARTIFACTS_DIR = PROJECT_ROOT / 'artifacts'
path = ARTIFACTS_DIR / 'training_history.csv'
if path.exists():
    history = pd.read_csv(path)
    print(history[['epoch', 'train_loss', 'val_loss', 'val_acc', 'val_f1_macro']].head(10).to_string(index=False))
else:
    print('No hay training_history.csv aún')
```
